Lets understand about a compiler and an interpretor to gain skill about scripting language

 **What is a Compiler**

 <b><br>
 1. Compiler is a software that converts a program written in high level language (Source Language) to low level language (Object/Target/Machine Language)
 2. They usually run on Computers with same architecture.
 3. Since they are directly converted into machine code they are faster.
 4. They generate an intermediate code inorder to revert back the object prioe state before deformation. <B>




 **What is an Interpreter**

 1. Makes the code understandable to the Machine's processor without
 converting into byte code.
 2. It executes the instructions line by line.
 3. They are processed during the runtime.
 4. They do not require the intermediate object hence memory effiecient
 Example : Java | Python
