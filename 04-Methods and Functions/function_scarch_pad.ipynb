```python
# Print even number between 10 to 100
# if the number is odd return 'Number is Odd'
# If even return 'Number is even'
# if the number is neutral 'return Neutral'
def check_odd_or_even(num1):
    if num1 ==0:
        return 'Number is Neutral'
    elif num1%2 ==0:
        return 'Number is Even'
    else:
        return 'Number is odd'
```

```python
for i in range(0,100):
    print(i, check_odd_or_even(i))
```

```python
# Create to take two string first and last name and return full name
def concat_strings(f_name,l_name):
    l_name = l_name.upper()
    return f_name+' '+l_name
```

```python
first_name = 'Nishanth'
last_name = 'rajendran'
print(concat_strings(f_name=first_name,l_name=last_name))
```
