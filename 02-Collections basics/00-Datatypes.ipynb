### Data Types in Python

1. **Text Type:**
   - `str`: Represents a string of characters, such as "Hello, World!".

2. **Numeric Types:**
   - `int`: Represents integers, e.g., 5, -12, 0.
   - `float`: Represents floating-point numbers, e.g., 3.14, -0.5.
   - `complex`: Represents complex numbers, e.g., 2 + 3j.

3. **Sequence Types:**
   - `list`: Represents an ordered collection of items, mutable (modifiable).
   - `tuple`: Represents an ordered collection of items, immutable (cannot be modified).
   - `range`: Represents a sequence of numbers within a specified range.

4. **Mapping Type:**
   - `dict`: Represents a collection of key-value pairs, where each key must be unique.

5. **Set Types:**
   - `set`: Represents an unordered collection of unique items. Duplicate values are not allowed.
   - `frozenset`: Similar to a set but is immutable.

6. **Boolean Type:**
   - `bool`: Represents boolean values, either `True` or `False`.

7. **Binary Types:**
   - `bytes`: Represents a sequence of bytes, immutable.
   - `bytearray`: Represents a mutable sequence of bytes.
   - `memoryview`: Provides a view on the memory of an object, which can be a sequence of bytes.

8. **None Type:**
   - `NoneType`: Represents the absence of a value or a null value.

These data types are fundamental in Python programming and are used to represent different kinds of values and structures in a program.


# Numeric  Type

````python
# Integer
import sys
a = 10
print("Type of Variable a ",type(a))
print('Memory Location of variable a',hex(id(a)))
print("Size of Variable a ",sys.getsizeof(a))
````

````python
#Float

import sys

f = 3.14
print("Type of Variable f:", type(f))
print('Memory Location of variable f:', hex(id(f)))
print("Size of Variable f:", sys.getsizeof(f))
````

````python
# Complex
import sys

complex_number = 2 + 3j
print("Type of Variable complex_number:", type(complex_number))
print('Memory Location of variable complex_number:', hex(id(complex_number)))
print("Size of Variable complex_number:", sys.getsizeof(complex_number))
````

# Sequence Type

### List

````python
import sys

# List
my_list = [1, 2, 3]
print("Type of Variable my_list:", type(my_list))
print("Memory Locations of values in my_list:")
for item in my_list:
    print(f"  - {item}: {hex(id(item))}")
print("Size of Variable my_list:", sys.getsizeof(my_list))
````

### **Tuple**

````python
# Tuple
my_tuple = (4, 5, 6)
print("Type of Variable my_tuple:", type(my_tuple))
print("Memory Locations of values in my_tuple:")
for item in my_tuple:
    print(f"  - {item}: {hex(id(item))}")
print("Size of Variable my_tuple:", sys.getsizeof(my_tuple))
````

### **Range**

````python
my_range = range(1, 4)
print("Type of Variable my_range:", type(my_range))
# Note: range objects don't have individual memory locations for each element as they are generated on the fly.
print("Size of Variable my_range:", sys.getsizeof(my_range))
````

# Mapping Type


````python
import sys

# Dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3}
print("Type of Variable my_dict:", type(my_dict))
print("Memory Locations of values in my_dict:")
for key, value in my_dict.items():
    print(f"  - {key}: {hex(id(value))}")
print("Size of Variable my_dict:", sys.getsizeof(my_dict))
````

# Muttable Vs Immuatable

| Characteristic       | Mutable Objects                         | Immutable Objects                       |
|-----------------------|-----------------------------------------|-----------------------------------------|
| **Definition**       | Objects whose state can be changed after creation. | Objects whose state cannot be changed after creation. |
| **Examples**         | Lists, dictionaries, sets, byte arrays, etc. | Tuples, strings, integers, floats, etc.  |
| **Modifiability**    | You can modify the content after creation (e.g., add or remove elements). | Once created, the content cannot be modified. |
| **Memory Efficiency** | May be less memory-efficient due to the need for extra space for potential modifications. | Generally more memory-efficient as their size remains constant. |
| **Syntax Example**   | ```python my_list = [1, 2, 3] my_list.append(4) ``` | ```python my_tuple = (1, 2, 3) # Attempting to modify will raise an error # my_tuple[0] = 0  # This line will result in an error ``` |


# **List vs tuple vs set vs dict**

---



| Characteristic       | List                               | Tuple                              | Set                                | Dictionary                         |
|-----------------------|------------------------------------|------------------------------------|------------------------------------|------------------------------------|
| **Definition**       | Ordered collection of elements.    | Ordered, immutable collection.    | Unordered collection of unique elements. | Unordered collection of key-value pairs. |
| **Mutability**       | Mutable; elements can be modified after creation. | Immutable; elements cannot be modified after creation. | Mutable; elements can be added or removed after creation. | Mutable; key-value pairs can be added, removed, or modified after creation. |
| **Representation**   | Represented by square brackets: `my_list = [1, 'two', 3.0]` | Represented by parentheses: `my_tuple = (1, 'two', 3.0)` | Represented by curly braces: `my_set = {1, 'two', 3.0}` | Represented by curly braces with key-value pairs: `my_dict = {'one': 1, 'two': 2, 'three': 3}` |
| **Example**          | `my_list = [1, 'two', 3.0]`        | `my_tuple = (1, 'two', 3.0)`       | `my_set = {1, 'two', 3.0}`         | `my_dict = {'one': 1, 'two': 2, 'three': 3}` |
| **Indexing**         | Supports indexing and slicing.     | Supports indexing and slicing.     | Does not support indexing or slicing. | Does not support indexing or slicing. |
| **Use Case**         | When ordered elements with duplicates are needed and modification is required. | When an ordered, immutable collection is needed. | When an unordered collection of unique elements is needed. | When key-value pairs are needed for quick lookups and modifications. |
